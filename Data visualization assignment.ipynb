```python
#Importing libraries and dependancies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
```

```python
#Reading the dataset
#Loading iris dataset from sklearn
data = datasets.load_iris()
#Extracting the data array from data dictionary
X = pd.DataFrame(data=data['data'], columns=data['feature_names'])
X.head()
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)

#Reducing the dimensions for 3D
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
X = pca.fit_transform(X)
```

```python
x_data = np.array([X[:,0]]).reshape(-1,1)
y_data = np.array([X[:,1]]).reshape(-1,1)
z_data = np.array([X[:,2]]).reshape(-1,1)
```

```python
fig = plt.figure(figsize=(10,10))
ax=plt.axes(projection='3d')
ax.scatter(x_data, y_data, z_data, c=z_data, cmap='viridis', linewidth=0.5)
plt.show()
```
